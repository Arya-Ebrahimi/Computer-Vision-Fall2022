```python
def show_img(*args, figsize=10, is_gray=True, title=None, fontsize=12):
    if isinstance(figsize, int):
        figsize = (figsize, figsize)
    images = args[0] if type(args[0]) is list else list(args)
    cmap=None
    if not is_gray:
        images = list(map(lambda x: cv2.cvtColor(x, cv2.COLOR_BGR2RGB), images))
    else:
        cmap = 'gray'
    plt.figure(figsize=figsize)
    for i in range(1, len(images)+1):
        plt.subplot(1, len(images), i)
        if title is not None:
            plt.title(title[i-1], fontsize=fontsize)
        
        plt.imshow(images[i-1], cmap=cmap)
        plt.axis('off')
```

```python
gaussian_kernel = 1/16 * np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])

def pixel_replication(img, k):
    res = np.repeat(img, k, axis=0)
    res = np.repeat(res, k, axis=1)
    return res

def average_downsample(img, filter_size):
    filter = np.ones([filter_size, filter_size])
    filter /= filter_size**2

    r = np.zeros((math.floor(img.shape[0]/filter_size), math.floor(img.shape[1]/filter_size)))
    for i in range(math.floor(img.shape[0]/filter_size)):
        for j in range(math.floor(img.shape[1]/filter_size)):
            mat = img[i*filter_size:(i+1)*filter_size, j*filter_size:(j+1)*filter_size]
            mul = mat * filter
            sum = np.sum(mul)
            r[i, j] = sum

    return r

def remove_row_col(img, k):
    return img[::k,::k]

    
def pad(img):
    shape_0 = img.shape[0]
    shape_1 = img.shape[1]

    if(img.shape[0]%2):
        shape_0 = img.shape[0]+1
    if(img.shape[1]%2):
        shape_1 = img.shape[1]+1

    y = np.zeros((shape_0, shape_1))
    y[0:img.shape[0], 0:img.shape[1]] = img
    return y
```

```python
def pyramid(img, level, kernel=gaussian_kernel, upsample_function=pixel_replication, downsample_function=remove_row_col):
    img = pad(img)
    G = np.zeros((img.shape[0], int(1.5*img.shape[1])))
    L = np.zeros((img.shape[0], int(1.5*img.shape[1])))

    g = []
    l = []

    G[0:img.shape[0], 0:img.shape[1]] = img
    g.append(img)

    imgcol = img.shape[1]
    pointer_g = 0
    pointer_l = 0

    for i in range(level):
        img = pad(img)
        kernelized = cv2.filter2D(img, -1, kernel)
        downsampled = downsample_function(kernelized, 2)
        upsampled = upsample_function(downsampled, 2)
        if (i == 0):
            L[:, 0:img.shape[1]] = img-upsampled
        else:
            L[pointer_l:pointer_l+img.shape[0], imgcol:imgcol+img.shape[1]] = img-upsampled
            pointer_l += img.shape[0]
        l.append(img-upsampled)
        img = downsampled
        G[pointer_g:pointer_g+img.shape[0], imgcol:imgcol+img.shape[1]] = img
        g.append(img)
        if (i == level-1):
            L[pointer_g:pointer_g+img.shape[0], imgcol:imgcol+img.shape[1]] = img
            l.append(img)
        pointer_g += img.shape[0]
        
    return G, L, g, l
```

```python
def rec_pyr(g, l, upsample_function):
    n = len(l)-1
    gg = g[-1]
    for i in range(n, 1, -1):
        upsampled = upsample_function(gg, 2)
        gg = upsampled + l[i-1]
    
    return gg
```

```python
from pywt import dwt2, idwt2

def wavelet(img, level, mode='haar'):

    Wrep = np.zeros((img.shape[0], img.shape[1]))
    LLrep = np.zeros((img.shape[0], int(1.5*img.shape[1])))
    LLrep[0:img.shape[0], 0:img.shape[1]] = img
    img_width = img.shape[1]
    pointer = 0
    end = (img.shape[0], img.shape[1])
    coeffs = {}

    for i in range(level):
        LL, (LH, HL, HH) = dwt2(img, 'haar')

        coeffs[i] = (LH, HL, HH)

        Wrep[0:LL.shape[0], 0:LL.shape[1]] = LL
        Wrep[0:LH.shape[0], LL.shape[1]:end[1]] = LH
        Wrep[LL.shape[0]:end[0], 0:HL.shape[1]] = HL
        Wrep[LL.shape[0]:end[0], LL.shape[1]:end[1]] = HH

        LLrep[pointer:pointer + LL.shape[0], img_width:img_width + LL.shape[1]] = LL/(2**(i+1))
        pointer += LL.shape[0]

        end = (int(end[0]/2), int(end[1]/2))
        img = LL
    
    return Wrep, LLrep, coeffs, LL
```

```python
def quantize_coeffs(coeffs, gamma=2):
    def formula(x):
        return gamma * np.sign(x) * np.floor(abs(x/gamma))

    for i in range(len(coeffs)):
        lh, hl, hh = coeffs[i]
        coeffs[i] = (formula(lh), formula(hl), formula(hh))
    
    return coeffs
```

```python
coeffs_qunatized = quantize_coeffs(coeffs, gamma=2)
```

```python
def rec(ll, coeffs):
    for i in range(len(coeffs)-1, -1, -1):
        ll = idwt2([ll, coeffs[i]], 'haar')

    return ll
```
